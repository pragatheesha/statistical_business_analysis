```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Sales": [42000, 45000, 48000, 50000, 52000, 55000, 58000, 60000],
    "Marketing_Spend": [8000, 9000, 10000, 11000, 12000, 13000, 14000, 15000],
    "Region": ["North", "South", "North", "East", "South", "North", "East", "South"]
}

df = pd.DataFrame(data)
df
```

```python
# Calculate Comprehensive Statistics for Business Data

df.describe()
```

```python
# manual statistics
mean_sales = df["Sales"].mean()
median_sales = df["Sales"].median()
std_sales = df["Sales"].std()

print("Mean:", mean_sales)
print("Median:", median_sales)
print("Standard Deviation:", std_sales)
```

```python
# Analyze Data Distribution (Histogram + Density)
sns.histplot(df["Sales"], kde=True)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Perform t-tests to Compare Groups
# Is sales different between North & South regions?
north_sales = df[df["Region"] == "North"]["Sales"]
south_sales = df[df["Region"] == "South"]["Sales"]

t_stat, p_value = stats.ttest_ind(north_sales, south_sales)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
# Calculate Correlation Between Business Metrics
correlation = df["Sales"].corr(df["Marketing_Spend"])
print("Correlation:", correlation)
```

```python
# visualize correation
sns.heatmap(df[["Sales", "Marketing_Spend"]].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Interpret p-values in Business Context
print("P-value:", p_value)
```

```python
# Calculate Confidence Intervals (95%)
mean = df["Sales"].mean()
sem = stats.sem(df["Sales"])

confidence_interval = stats.t.interval(
    0.95,
    len(df["Sales"]) - 1,
    mean,
    sem
)

print("95% Confidence Interval:", confidence_interval)
```
